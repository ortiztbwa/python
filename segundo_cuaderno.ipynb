```python
import pandas as pd
import numpy as np

#create dataframe from lists
Nombre = ['Sebastian Ortiz', 'Viviana Kanagusku', 'Irma Kanagusku']
Edad = [39, 40, 84]
df = pd.DataFrame({
  'Nombre': Nombre,
  'Edad': Edad,
})
df
```

```python
#create dataframe from dict
personas = {
    "Nombre":["Norma Romero", "Guillermo Ortiz", "Gabriela Ortiz"],
    "Edad":[59,60,41]
}
df2 = pd.DataFrame.from_dict(personas)
df2
```

```python
#concat dataframes
df_f = pd.concat([df,df2],axis=0)
df_f.reset_index(inplace = True)
df_f.drop(["index"],axis=1, inplace = True)
df_f.set_index(keys="Nombre", inplace = True)
df_f
```

```python
#merge dataframes
mas_datos = {
    "Nombre":['Sebastian Ortiz', 'Viviana Kanagusku', 'Irma Kanagusku',"Norma Romero", "Guillermo Ortiz", "Gabriela Ortiz"],
    "Altura":[1.7,1.6,1.5,1.6,1.65,1.65],
    "Genero":["M","F","F","F","M","F"]
}
df3 = pd.DataFrame.from_dict(mas_datos)
df3.set_index(keys="Nombre", inplace = True)
df3
df_f = df_f.merge(df3, left_on="Nombre", right_on="Nombre")
df_f
```

```python
#updates dtypes
df_f.dtypes
```

```python
#apply function
df_f["Enano"]=True
def enano(col):
  Altura = col[0]
  Enano = col[1]
  if Altura>1.6:
    return False
  else:
    return np.nan

df_f["Enano"] =df_f[["Altura","Enano"]].apply(enano, axis=1)
df_f  
```

```python
#na treatment
#df_f = df_f.dropna(axis=1,how='all')
#df_f = df_f.dropna()
df_f["Enano"].replace(np.nan, True, inplace=True)

df_f
```
