```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

url = "https://en.wikipedia.org/wiki/List_of_S%26P_500_companies"
html = requests.get(url).content
soup = BeautifulSoup(html)

table = soup.find("table", {"id": "constituents"})
df = pd.read_html(str(table))[0]
df
```

```python
from numpy.core import numeric
df.dtypes
df["Founded"] = df["Founded"].astype(str)
df["Date first added"] = df["Date first added"].astype(str)
df["Founded"] = df["Founded"].str.slice(0, 4)
df["Date first added"] = df["Date first added"].str.slice(0, 10)
df["Founded"] = pd.to_datetime(df["Founded"])
df["Date first added"] = pd.to_datetime(df["Date first added"])

df["year_diff"] = (df["Date first added"] - df["Founded"]).dt.days // 365.25
df["year_diff"].fillna(0, inplace=True)
df["year_diff"] = df["year_diff"].astype(int)
df
```
